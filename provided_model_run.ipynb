```python
from ultralytics import YOLO
import cv2
import os
from tqdm import tqdm
import time

# Load YOLOv8 model
model = YOLO("Th_G_v9.pt")

# Corrected input path
input_dir = "data/test/grayscale_images"
output_dir = "outputs"
os.makedirs(output_dir, exist_ok=True)

# List all image files
image_files = [f for f in os.listdir(input_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

if not image_files:
    print("No image files found in:", input_dir)

for filename in image_files:
    image_path = os.path.join(input_dir, filename)
    # print(f"Running inference on {image_path}...")

    # Run the model
    results = model(image_path, save=False)

    for result in results:
        annotated_img = result.plot()  # numpy array with detections
        save_path = os.path.join(output_dir, f"pred_{filename}")
        cv2.imwrite(save_path, annotated_img)
        # print(f"Saved: {save_path}")


```

```python
from tqdm import tqdm
import os
import cv2

input_folder = "data/test/images"
output_folder = "data/test/grayscale_images"
os.makedirs(output_folder, exist_ok=True)

image_files = [f for f in os.listdir(input_folder) if f.lower().endswith((".jpeg", ".jpg", ".png"))]

for filename in tqdm(image_files, desc="Converting to Grayscale"):
    img_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)

    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(output_path, gray)
```
